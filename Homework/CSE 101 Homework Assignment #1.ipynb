#### Part 1: Evaluate a Mathematical Formula (10 points)

Complete the function `evaluate`, which returns the value of the mathematical function $f(x,y) = \frac{xy}{y+2}-\frac{y}{x}$. Don't use integer division!

Expected Results:

Function Call | Return Value
:-- | :--
`evaluate(-2, 0)` | `0.0`
`evaluate(-5, -5)` | `-9.333333333333334`
`evaluate(4, -4)` | `9.0`
`evaluate(-1, 5)` | `4.285714285714286`
`evaluate(-1, 2)` | `1.5`

```
def evaluate(x, y):
    return (x*y/(y+2)) - (y/x) 


# Test cases
print(evaluate(-2, 0))
print(evaluate(-5, -5))
print(evaluate(4, -4))
print(evaluate(-1, 5))
print(evaluate(-1, 2))
```

#### Part 2: Gas Refill Cost Calculator (10 points)

Complete the function `gas_cost`, which returns how much it costs in fuel to drive a car a given distance. The function takes a `distance` in miles a car has been driven, the fuel efficiency in `mpg` (miles per gallon), and the `price` of a gallon of gass.

Expected Results:

Function Call | Return Value
:-- | :--
`gas_cost(59, 38.8, 3.33)` | `5.063659793814433`
`gas_cost(14, 21.1, 3.98)` | `2.6407582938388625`

```
def gas_cost(distance, mpg, price):
    return distance*price/mpg


# Test cases
print(gas_cost(59, 38.8, 3.33))
print(gas_cost(14, 21.1, 3.98))
print(gas_cost(29, 27.6, 4.22))
print(gas_cost(69, 32.3, 3.82))
print(gas_cost(195, 38.0, 4.40))
```

#### Part 3: To the Moon! (20 points)

Complete the function `rocket_area`, which returns the surface area of the rocket ship depicted below. The function arguments are:

* `c`: the height of the cone
* `r`: the radius of the bottom of the cone (and also the cylinder)
* `h`: the height of the cylinder
* `b`: the length of the sides of the wings/fins on the rocket (which are four isosceles, right triangles)

Expected Results:

Function Call | Return Value
:-- | :--
`rocket_area(19.2, 40.12, 23.39, 12.03)` | `16848.362563864128`
`rocket_area(45.5, 11.78, 33.83, 44.12)` | `8572.442691008864`
`rocket_area(14.32, 22.23, 37.67, 20.86)` | `9531.054099270114`
`rocket_area(29.82, 48.1, 31.23, 46.9)` | `29657.91102812789`
`rocket_area(48.58, 40.9, 38.94, 40.44)` | `26692.73526068156`

```
from IPython.display import display, HTML
display(HTML('''<img src="https://www.cs.stonybrook.edu/~ktm/courses/cse101/colab_images/rocket.png" height="300">'''))
```

```
import math

def nose_area(c,r):
    return math.pi*r*(math.sqrt(c**2+r**2))

def body_area(r,h):
    return 2*math.pi*r*h + math.pi*r**2

def wing_area(b):
    return (b**2)/2

def rocket_area(c, r, h, b):
    return nose_area(c,r) + body_area(r,h) + 4*wing_area(b)

# Test cases
print(rocket_area(19.2, 40.12, 23.39, 12.03))
print(rocket_area(45.5, 11.78, 33.83, 44.12))
print(rocket_area(14.32, 22.23, 37.67, 20.86))
print(rocket_area(29.82, 48.1, 31.23, 46.9))
print(rocket_area(48.58, 40.9, 38.94, 40.44))
```
