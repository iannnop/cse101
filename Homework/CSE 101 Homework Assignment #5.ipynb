### Part 1: One-armed Bandit (20 points)

The Wolfie Slot Machine contains three reels, each of which randomly displays one of `'Cherry'`, '`Plum'`, '`Apple'` or `'Orange`' when spun. Each of the four outcomes has an equal chance of appearing (i.e., 25%). On each pull of the slot machine's arm (which costs \$1), the player wins or loses money according to the following rules:
* 3 Cherries earn the player \$25
* 2 Cherries earn the player \$4
* 1 Cherry earns the player \$1
* all other results *cost* the player the \$1 that was inserted

Suppose the player pulls the level and 2 Cherries appear. Then the player wins \$4 - \$1 = \$3. Getting 1 Cherry would cause the player to net no money: \$1 - \$1 = \$0.

**Part 1(a)**

Write code that simulates someone using the slot machine. The player starts with $5000 and pulls the lever 1000 times. At the end of the simulation, print the ending balance. *For this part of the assignment you do not need to write a function.*

````
#Part 1(a)

import random

funds = 5000
for i in range(1000):
    funds -= 1
    roll = []
    for j in range(3):
        slot = random.random()
        if slot <= 0.25:
            roll.append("Cherry")
        elif slot <= 0.50:
            roll.append("Plum")
        elif slot <= 0.75:
            roll.append("Apple")
        else:
            roll.append("Orange")
    cherry = roll.count("Cherry")
    if cherry == 3:
        funds += 25
    else:
        funds += cherry ** 2
print(funds)
````

### Part 2: The Tortoise and the Hare (20 points)

Simulate the race between the tortoise and the hare from the famous children's story. In our version, the two animals race down a path that is 100 meters long in an effort to be the first to reach the finish line. *For this part of the assignment you do not need to write a function.*

During each minute, the tortoise has a:
* 25% probability of not moving at all
* 30% probability of moving forward 1 meter
* 35% probability of moving forward 2 meters
* 10% probability of moving forward 3 meters

During each minute, the hare has a:
* 10% probability of moving *backward* 1 meter, possibly moving it before the starting line
* 65% probability of not moving at all
* 25% probability of moving forward 5 meters

Simulate the race until one of the animals passes the finish line. A suggestion: for each animal, keep track of its distance from the starting line. When either or both have reached a distance of 100 meters or greater from the starting line, the race is over. 

At the end of the simulation, print the distances traveled by the tortoise and hare, and also which was the winner (or if it was a tie, which happens when both animals achieve a distance of 100 or greater in the same iteration of your simulation loop).

```

```

Hare distance: 102

````
import random

tortoise = 0
hare = 0

while tortoise < 100 and hare < 100:
    tortoise_roll = random.random()
    hare_roll = random.random()
    
    if tortoise_roll <= 0.25:
        pass
    elif tortoise_roll <= 0.55:
        tortoise += 1
    elif tortoise_roll <= 0.90:
        tortoise += 2
    else:
        tortoise += 3
        
    if hare_roll <= 0.10:
        hare -= 1
    elif hare_roll <= 0.75:
        pass
    else:
        hare += 5

if tortoise >= 100 and hare >= 100:
    result = "Tie!"
else:
    result = "Tortoise wins!" if tortoise >= 100 else "Hare wins!"

print(f"Tortoise distance: {tortoise}\nHare distance: {hare}\n{result}")
````

### Part 3: Analyze the Tortoise/Hare Race (20 points)

Now we want to analyze how often the tortoise and hare win or lose the race as we vary the race length. We can wrap the logic of our race simulation in a function and then call it many times to simulate many races. Then, we can collect some statistics to see how likely is it that the tortoise (or hare) wins a race of a given length.

Specifically, write a function `simulate_race` that takes one argument, `race_length`, which is a positive integer that gives the length of the race (so, 100 is no longer-hard coded as the race's length). The function returns one of three strings: '`tortoise'`, `'hare'` or `'tie'`, to indicate which animal won the race or whether there was a tie.

Outside your `simulate_race` function, in a loop, call your method exactly 10000 times with varying race lengths as indicated in the table of experiments below and compute the winning percentages of the tortoise and hare, and also the percentage of times that they tie. What you will find is that the percentages vary as the race length changes.

Below is the algorithm you should implement for running the experiments:
```
initialization of variables to compute statistics
for i in range(10000):
    call simulate_race(N)  # where N is 50, 100, 250 or 1000 for this experiment
    update statistics variables as needed
print winning % for tortoise and hare (separately), and % ties for this experiment
```

Expected Results (all percentages are approximate):

Experiment | % Tortoise Wins | % Hare Wins | % Ties
:-- | :-- | :-- | :--
10,000 calls of `simulate_race(50)` | 63% | 3% | 34%
10,000 calls of `simulate_race(100)` | 70% | 2% | 28%
10,000 calls of `simulate_race(250)` | 79% | 1% | 20%
10,000 calls of `simulate_race(1000)` |95% | 0% | 5%

*Slow and steady* really does win the race, especially if it's a long race!

````
import random


def simulate_race(race_length):
    tortoise = 0
    hare = 0

    while tortoise < race_length and hare < race_length:
        tortoise_roll = random.random()
        hare_roll = random.random()

        if tortoise_roll <= 0.25:
            pass
        elif tortoise_roll <= 0.55:
            tortoise += 1
        elif tortoise_roll <= 0.90:
            tortoise += 2
        else:
            tortoise += 3

        if hare_roll <= 0.10:
            hare -= 1
        elif hare_roll <= 0.75:
            pass
        else:
            hare += 5

    if (tortoise >= race_length) and (hare >= race_length):
        result = "tie"
    else:
        result = "tortoise" if tortoise >= race_length else "hare"

    return result


race50 = {"tortoise": 0, "hare": 0, "tie": 0}
race100 = {"tortoise": 0, "hare": 0, "tie": 0}
race250 = {"tortoise": 0, "hare": 0, "tie": 0}
race1000 = {"tortoise": 0, "hare": 0, "tie": 0}
cases = 10000

for i in range(cases):
    race50[simulate_race(50)] += 1
    race100[simulate_race(100)] += 1
    race250[simulate_race(250)] += 1
    race1000[simulate_race(1000)] += 1

for result in race50:
    race50[result] /= 100
    race100[result] /= 100
    race250[result] /= 100
    race1000[result] /= 100

print(
    f"""
-------------------------------------------------------
| Experiment | % Tortoise Wins | % Hare Wins | % Ties |
-------------------------------------------------------
|     50     |      {race50['tortoise']:.2f}%     |    {race50['hare']:.2f}%   | {race50['tie']:.2f}%  |
|     100    |      {race100['tortoise']:.2f}%     |    {race100['hare']:.2f}%   | {race100['tie']:.2f}%  |
|     250    |      {race250['tortoise']:.2f}%     |    {race250['hare']:.2f}%   | {race250['tie']:.2f}%  |
|     1000   |      {race1000['tortoise']:.2f}%     |    {race1000['hare']:.2f}%    | {race1000['tie']:.2f}%  |
-------------------------------------------------------
"""
)
````

### Part 4: Identifying Poker Hands

Write the function `poker_hand` that takes a list of exactly five distinct `Card` objects as an argument, analyzes the list, and returns one of the following strings that describes the hand:

* `'Four of a kind'` (four cards of the same rank)
* `'Full house'` (three cards of one rank, and two cards of a different rank)
* `'Flush'` (five cards of the same suit)
* `'Three of a kind'` (exactly three cards of the same rank)
* `'One pair'` (at least one pair of the same rank, but not four cards of the same rank)
* `'High card'` (no cards of identical rank that also do not form a straight)

You only need to identify the above Poker hands. 

````
import random

class Card:
    suits = {'C': '\u2663', 'D': '\u2666', 'H': '\u2665', 'S': '\u2660'}

    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __str__(self):
        return self.rank + Card.suits[self.suit]
    
    def __repr__(self):
        return f"Card('{self.rank}','{self.suit}')"

def generate_deck():
    deck = [Card(r, s) for r in 'A23456789TJQK' for s in 'CDSH']
    random.shuffle(deck)
    return deck

def poker_hand(cards):
    known_suits = {}
    known_ranks = {}
    for card in cards:
        if card.suit in known_suits:
            known_suits[card.suit] += 1
        else:
            known_suits[card.suit] = 1

        if card.rank in known_ranks:
            known_ranks[card.rank] += 1
        else:
            known_ranks[card.rank] = 1

    if any([known_ranks[rank] == 4 for rank in known_ranks]):
        return "Four of a Kind"
    elif any([known_ranks[rank] == 3 for rank in known_ranks]) and any([known_ranks[rank] == 2 for rank in known_ranks]):
        return "Full House"
    elif any([known_suits[suit] == 5 for suit in known_suits]):
        return "Flush"
    elif any([known_ranks[rank] == 3 for rank in known_ranks]):
        return "Three of a Kind"
    elif any([known_ranks[rank] == 2 for rank in known_ranks]):
        return "One Pair"
    else:
        return "High Card"


# Use code like the following to generate random 5-card hands!
# hand = generate_deck()[:5]

# Test cases
four_of_a_kind = [Card('6','H'), Card('6','S'), Card('6','C'), Card('3','H'), Card('6','D')]
full_house = [Card('J','C'), Card('J','S'), Card('T','C'), Card('J','H'), Card('T','S')]
flush = [Card('K','S'), Card('4','S'), Card('Q','S'), Card('6','S'), Card('5','S')]
three_of_a_kind = [Card('2','C'), Card('6','S'), Card('6','C'), Card('3','H'), Card('6','D')]
one_pair = [Card('J','D'), Card('4','C'), Card('5','H'), Card('Q','H'), Card('4','H')]
high_card = [Card('7','D'), Card('3','D'), Card('K','C'), Card('9','D'), Card('5','S')]
hand = generate_deck()[:5]
print(poker_hand(four_of_a_kind))
print(poker_hand(full_house))
print(poker_hand(flush))
print(poker_hand(three_of_a_kind))
print(poker_hand(one_pair))
print(poker_hand(high_card))
print(poker_hand(hand), hand)
````
