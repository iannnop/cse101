### Part 1: Using "Vanilla" Python, implement the formula $z = \dfrac{\sum_{i=1}^n (x_i + y_i)^{3.1}}{5 \sum_{i=1}^n (4x_i - 2y_i)}$ $\quad$ (5 points)

A test case has been provided below as the `x` and `y` lists. The expected value of $z$ for these inputs is $16.65459061574281$. Be sure to display your computed value!

**Do not use NumPy to complete this part.**

```
x = [8, 6, 7, 5, 3, 1, 9, 5, 6]
y = [7, 1, 3, 0, 9, 4, 3, 2, 1]

z = sum([(xi+yi)**3.1 for xi, yi in zip(x,y)])/(5*sum([(4*xi-2*yi) for xi, yi in zip(x,y)]))
z
```

### Part 2: Using NumPy, implement the formula $z = \dfrac{\sum_{i=1}^n (x_i + y_i)^{3.1}}{5 \sum_{i=1}^n (4x_i - 2y_i)}$ $\quad$ (10 points)

(4 points) Task 10: Now, transpose and visualize the *original* dataset so that along the horizontal axis we have *all* age groups and along the vertical axis we show the total time spent. Set the title and axis labels as in the example below. Move the legend outside the figure. 

```
from IPython.display import display, HTML
display(HTML('''<img src="https://www.cs.stonybrook.edu/~ktm/courses/cse101/colab_images/homework6d.png"">'''))
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

labels = df.columns[1:]
datasets = {}
datacount = len(df.values[:,0])
for i in range(datacount):
    datasets[f'data{i}'] = df.values[i,1:]

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots(figsize=(10,5))

for i in range(datacount):
    if i == 0:
        ax.bar(x, datasets[f'data{i}'], width, label=df.values[i,0])
        prev = datasets['data0']
    else:
        ax.bar(x, datasets[f'data{i}'], width, bottom=prev, label=df.values[i,0])
        prev += datasets[f'data{i}']

ax.set_xlabel('Age Groups')
ax.set_ylabel('Activity Distribution in Hours')
ax.set_title('Activities of People')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
