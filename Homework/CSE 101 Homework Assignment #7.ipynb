````
import numpy as np

def image_compress(file_name):
    colored = []
    with open(file_name) as image:
        line = [line.strip() for line in image]
    block = np.array([line[i].split(' ') for i in range(len(line))])

    for i in range(len(block)):
        for j in range(len(block[i])):
            if block[i,j] != '255':
                colored += [(i,j,int(block[i,j]))]

    return colored



# Test cases
print(image_compress('image1.txt'))
print(image_compress('image2.txt'))
print(image_compress('image3.txt'))
````

### Part 2: A Simple Text Compression Scheme (25 points)

In this problem, you will be implementing a basic text compression algorithm. Write a function `text_compress()`, which takes two arguments:

* `infile`: the name of the file whose contents we want to compress

* `outfile`: the name of the file to store the compressed result

Read the file line by line, compressing each line as described below. Append each compressed line into a list. At the end of processing, write the contents of the list to disk, one string per line. Add a newline at the end of each compressed line when writing it to disk. Assume that the input file contains no digit characters.

The compression algorithm works like this: take a line of text from the input file and start reading from the beginning, compressing identical consecutive characters in the form of a (i) count of that character directly followed by (ii) that character. In the end, the original line
will be shortened (hopefully!), and the function should return that shortened result.

For example, the line `'AAABBBCCC'` will be shortened to `'3A3B3C'`; the line `'AAC!!?????'` will be
shortened to `'2A1C2!5?'`; and the line `'____@@@@B__'` will be shortened to `'3_4@1B2_'`, where `_` denotes a space in this example. Notice that a space
is also considered a character.

Note: You may assume that only valid filenames will be given as an input for the function. Also, it is very important
that you do a `.strip('\n\r')` on every line of input text. This is because the invisible newline/return characters may
cause issues.

Input file #3: [`runs.txt`](https://drive.google.com/file/d/1LxoqtrXCHCt8Z0JWnOdMQpRmtEL819jc/view?usp=sharing)

```

ccccccccccc
ddddddddddd

```

Expected output:

```

11b
11c
11d

4g7h
4m23k
4w8o7l6f10i11e
```

````
def text_compress(infile, outfile):
    outlist = []
    with open(infile) as inputfile:
        lines = inputfile.readlines()
    lines = [line.strip('\n\r') for line in lines]
    for line in lines:
        outstring = ''
        for i in range(len(line)):
            if i == 0:
                current = line[i]
                count = 1
            elif i == len(line)-1:
                count += 1
                outstring += f'{count}{line[i]}'
            else:
                if line[i] == current:
                    count += 1
                else:
                    outstring += f'{count}{line[i-1]}'
                    current = line[i]
                    count = 1
        outlist.append(outstring)
    with open(outfile, 'w') as output:
        for line in outlist:
            output.write(line)
            if line != outlist[len(outlist)-1]:
                output.write('\n')

    print(f'Stored compression of \'{infile}\' in \'{outfile}\'')


# Test cases
text_compress('simple.txt', 'compressed_simple.txt')
text_compress('letters.txt', 'compressed_letters.txt')
text_compress('runs.txt', 'compressed_runs.txt')
````

### Part 3: A Simple Encryption Scheme [Encryption] (25 points)

Write a function `sbu_encrypt()` that takes two arguments:

* `plaintext`: a plaintext string containing letters, digits and spaces that will be encrypted

* `num_rows`: the number of rows that will be used in the encryption process


Here's how the cipher works. First, take a plaintext string and start drawing characters from it, placing the first `num_rows` of characters into the first column of a grid from top to bottom. We assume here that the grid has `num_rows` rows in it. Then, take the next `num_rows` characters, placing them into the second column of the grid in reverse order, from bottom to top. Continue in this fashion, drawing characters from the plaintext string until you run out of characters, alternating between placing them from up-to-down and down-to-up. With some arithmetic you should be able to figure out exactly how many columns your grid will need to store all the characters from `plaintext`.

As an example, suppose we want to encrypt the string `'ABCDEFGHIJKLMNOPQRSTUVW'` with `num_rows = 5`. The characters would be placed into a grid as follows (where `d` means we are processing the characters in the downward direction, and `u` means we are processing characters in the upward direction):

```
dudud
AJKTU
BILSV
CHMRW
DGNQ
EFOP
```

If all characters from the plaintext are placed into the grid, but the last column is incomplete, we fill the last column with `%` symbols:
```
AJKTU
BILSV
CHMRW
DGNQ%
EFOP%
```

Next, take each row of characters, from left to right, and concatenate them to generate the final, encrypted string. For the example above, the final encrypted string would be: `AJKTU + BILSV + CHMRW + DGNQ% + EFOP% = AJKTUBILSVCHMRWDGNQ%EFOP%`.

Here's another example:

```
plaintext = 'STONYBROOKUNIV'
num_rows = 4
```

Insert the characters into the grid:

```
SOO
TRK
OBUV
NYNI
```

Insert `%` symbols into the last column, as necessary:

```
SOO%
TRK%
OBUV
NYNI
```

Concatenate the rows of characters together to generate the final, encrypted string: `SOO%TRK%OBUVNYNI`.

Note: `num_rows` must be greater than zero. In the case  that `num_rows` $\leq 0$, your function should return `plaintext`. In the case that `plaintext` is an empty string, your function should return a string of `num_rows` `%` symbols.

Examples:

Function Call | Return Value
:-- | :--
`sbu_encrypt('AMERICA', 7)` | `AMERICA`
`sbu_encrypt('STONYBROOKUNIV', 4)` | `SOO%TRK%OBUVNYNI`
`sbu_encrypt('CHICKENWINGS', -2)` | `CHICKENWINGS`
`sbu_encrypt('', 5)` | `%%%%%`
`sbu_encrypt('StonyBrookUniversityNewYork', 8)` | `Srs%tei%ovt%niy%ynN%BUekrkwrooYo`
`sbu_encrypt('UnitedStatesOfAmerica', 2)` | `UtetasOmecanidStefAri%`



````
def sbu_encrypt(plaintext, num_rows):
    if num_rows <= 0:
        return plaintext
    elif plaintext == '':
        return num_rows*'%'
    
    row = 0
    step = 0
    plaintext += (len(plaintext) % num_rows)* '%'
    a = ['']*num_rows
    for c in plaintext:
        a[row] += c
        if row == 0:
            step += 1
        elif row == num_rows-1:
            step -= 1
        row+=step
    encrypted = ''.join(a)

    return encrypted


# Test cases
print(sbu_encrypt('AMERICA', 7))
print(sbu_encrypt('STONYBROOKUNIV', 4))
print(sbu_encrypt('CHICKENWINGS', -2))
print(sbu_encrypt('', 5))
print(sbu_encrypt('StonyBrookUniversityNewYork', 8))
print(sbu_encrypt('UnitedStatesOfAmerica', 2))
````

### Part 4: A Simple Encryption Scheme [Decryption] (25 points)

Write a function `sbu_decrypt()` that takes the following two arguments:

* `encrypted`: ciphertext that was generated from plaintext by using the cipher described in Part 3

* `num_rows`: the number of rows that was used to encrypt the original plaintext string


Your function should take the encrypted string, decrypt it by applying the encryption algorithm from Part 3 *backwards*, and then return the (decrypted) original plaintext.

Note: `num_rows` must be greater than zero. In the case that `num_rows` $\leq 0$, your function should return `encrypted`. In the case that `encrypted` is an empty string, your function should return the value `None`.

Function Call | Return Value
:-- | :--
`sbu_decrypt('AMERICA', 7)` | `AMERICA`
`sbu_decrypt('SOO%TRK%OBUVNYNI', 4)` | `STONYBROOKUNIV`
`sbu_decrypt('CHICKENWINGS', -2)` | `CHICKENWINGS`
`sbu_decrypt('', 5)` | `Non`
`sbu_decrypt('Srs%tei%ovt%niy%ynN%BUekrkwrooYo', 8)` | `StonyBrookUniversityNewYork`
`sbu_decrypt('UtetasOmecanidStefAri%', 2)` | `UnitedStatesOfAmerica`

````
import numpy as np

def sbu_decrypt(encrypted, num_rows):
    if num_rows <= 0:
        return encrypted
    elif encrypted == '':
        return None
    elif len(encrypted) <= num_rows:
        return encrypted
        
    num_columns = len(encrypted)//num_rows
    a = np.array(list(encrypted)).reshape(num_rows, num_columns)
    b = []
    for i in range(num_columns):
        if i % 2 != 0:
            b.append(np.flip(a[:,i]).tolist())
        else:
            b.append(a[:,i].tolist())
    b = [''.join(c) for c in b]
    b = ''.join(b)
    decrypted = b.replace('%','')
    
    return decrypted
    


# Test cases
print(sbu_decrypt('AMERICA', 7))
print(sbu_decrypt('SOO%TRK%OBUVNYNI', 4))
print(sbu_decrypt('CHICKENWINGS', -2))
print(sbu_decrypt('', 5))
print(sbu_decrypt('Srs%tei%ovt%niy%ynN%BUekrkwrooYo', 8))
print(sbu_decrypt('UtetasOmecanidStefAri%', 2))
````
