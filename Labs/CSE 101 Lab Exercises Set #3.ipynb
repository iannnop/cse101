### Iterating Over Lists

Define a variable called `scores` and store inside it 7 or 8 integers in the range 0 through 100. Include some values near 0 and some values near 100.

```
scores = [60, 50, 100, 0, 80, 90, 70]
```

Use a for-loop to compute the total of those values greater than 50. Display the sum. Do not use list comprehensions to solve this problem.

```
total = 0

for score in scores:
    if score > 50:
        total += score
total
```

Use a for-loop and the `enumerate` function to print the indices of those scores greater than 50. Do not use list comprehensions to solve this problem.

```
for i, score in enumerate(scores):
    if score > 50:
        print(i)
```

Create two lists: a list of 5 first names and a list of 5 last names. Using a for-loop and `zip`, print 5 full names produced by concatenating a first name and last name from each list. Pull each name (one first name, one last name) from the same position (index) from each list.

```
first_names = ['Julius', 'John', 'Mary', 'Daniel', 'Lance']
last_names = ['Caesar', 'Doe', 'Jane', 'Arciaga', 'Jeung']
full_names = []
for first, last in zip(first_names, last_names):
    full_names += [first + ' ' + last]
full_names
```

Using your `scores` list from above, use a list comprehension to generate a list of scores less than 65 and store the resulting list in a variable named `failing`.

```
failing = [score for score in scores if score < 65]
failing
```

Create a list called `titles` and store in it a list of 5 book titles.

### Logical Operators

Write a code that checks if the value stored in a variable `x` is divisible by 2 and, if so, print "Even"; otherwise, print "Odd". Write your own test cases to check the correctness of your code.

```
x = 3
print('Even') if x % 2 == 0 else print('Odd')
```

Write a code that checks if the value stored in a variable `x` is divisible by 2 or 7, and, if so, print "Divisible"; otherwise, print "Not divisible". Write your own test cases to check the correctness of your code.

```
x = 8
print('Divisible') if x % 2 == 0 or x % 7 == 0 else print('Not divisible')
```

Write a code that checks if the string stored in a variable `x` is at least 10 characters long and starts with the letter Q. If so, print "Pass"; otherwise, print "Fail". Write your own test cases to check the correctness of your code.

```
x = 'QUEEN OF HEARTS'
print('Pass') if len(x) >= 10 and x[0] == 'Q' else print('Fail')
```

Write a code that tests if the string stored in a variable `x` starts with the letters X, Y or Z. If so, print "Pass"; otherwise, print "Fail".

```
x = 'YELLOW'
print('Pass') if x[0] == 'X' or x[0] == 'Y' or x[0] == 'Z' else print('Fail')
```

Write a code that checks if the string stored in a variable `x` starts with the letters X, Y or Z and ends with the letter A. If so, print "Pass"; otherwise, print "Fail". Write your own test cases to check the correctness of your code.

```
x = 'ZEBRA'
print('Pass') if x[0] == 'X' or x[0] == 'Y' or x[0] == 'Z' and x[-1] == 'A' else print('Fail')
```

Write code containing a while-loop that prints the first 5 multiples of 4, starting with 4 itself.

```
i = 1
while i < 6:
    print(4*i)
    i += 1
```

Write code containing a while-loop that prints all values in the list `nums` that are less than or equal to 10 **and** divisible by 3. Write your own test cases to check the correctness of your code.



```
nums = [0, 1, 2, 3, 10, 12, 18, 15, 6, 24, 9]
i = 0
while i < len(nums):
    if nums[i] <= 10 and nums[i] % 3 == 0:
        print(nums[i])
    i += 1
```
