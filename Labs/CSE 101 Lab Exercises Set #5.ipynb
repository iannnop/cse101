### Classes and Objects

Run the cell below.

```
class Card:
    suits = {'C': '\u2663', 'D': '\u2666', 'H': '\u2665', 'S': '\u2660'}

    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __str__(self):
        return self.rank + Card.suits[self.suit]
    
    def __repr__(self):
        return self.__str__()
```
