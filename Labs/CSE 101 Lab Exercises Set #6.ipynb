### NumPy Array Slicing

Using `arange`, create an array called `a` that contains the values 1, 2, 3, ..., 20. Use slicing to display the even values in `a`.

```
a = np.arange(1,20)
a[a % 2 == 0]
```

Using `arange`, create an array called `a` that contains the values 1, 2, 3, ..., 20. Use slicing to display the values at indices 0, 3, 6, 9, 12, 15, 18.

```
a = np.arange(1,20)
a[:19:3]
```

Using any means you wish, create a 2D array of integers with 4 rows and 6 columns called `a`. Include a mixture of positive and negative values. Display `a`.

```
import random
numbers = list(range(-12,12))
a = np.array([list(random.sample(numbers,6)) for i in range(4)])
a
```

Using slicing, display column 3.

```
print(a[:,3])
```

Using slicing, display the elements from rows #0 and 1, and columns #2 through 5 (inclusive).

```
print(a[:2,2:6])
```

Using the slicing notation that can accept conditions, display the values in `a` that are multiples of 2 or multiples of 3.

```
a[(a % 2 == 0) | (a % 3 == 0)]
```

Using the slicing notation that can accept conditions, display the values in `a` that are less than 0.

```
a[a<0]
```

### NumPy Arithmetic

Using any means you wish, create an array of 6 integers called `salaries`. Then, add the constant 100 to each value using NumPy syntax. This should require only one line of code. Display the array resulting from this addition.

```
salaries = np.array([22,12,370,582,10,65])
salaries += 100
salaries
```

Using any means you wish, create a 2D array of integers with 4 rows and 6 columns called `a`. Include a mixture of positive and negative values. Display the sum of the values in row 1.

```
import random
numbers = list(range(-12,12))
a = np.array([list(random.sample(numbers,6)) for i in range(4)])
a
```

Display the sum of the values in column 4.

```
a[:,4].sum(axis=0)
```

Display the mean of the values in column 3.

```
a[:,3].mean(axis=0)
```

Display the minimum value in row 2.

```
a[2,:].min()
```
