Write a list comprehension that creates a list called `b` that contains the square of every value from the list `a` given below.

Below are two lists that contain the names of 10 students and their respective ages. Write a dictionary comprehension that creates a dictionary called `people` that maps a person's name to his/her age. For example, `people['Janet']` should be mapped to `21`, `people['Bill']` should be mapped to `19`, and so on.

```
names = ['Janet', 'Bill', 'Amanda', 'Mike', 'Susan', 'Jen', 'Sara', 'Maria', 'Kathy', 'Andrew']
ages = [21, 19, 22, 25, 24, 29, 30, 18, 20, 27]
people = {name:ages for name, ages in zip(names, ages)}
people
```

Write a dictionary comprehension that creates a dictionary called `reversal` which provides the reverse mapping provided by your `people` dictionary. For example `reversal[21]` should be mapped to `'Janet'`. **Create `reversal` directly from `people`; do not create the new dictionary from the `names` and `ages` lists.**

```
reversal = {age:name for name, age in people.items()}
reversal
```

### Improve the Vigenère Cipher

Below is given the code for the Vigenère Cipher's encryption alorithm. Improve it by supporting lowercase letters in the keyword and plaintext message. Assume that the keyword and plaintext will contain only letter characters. Follow the model of the simpler substitution ciphers to incorporate this change.

Examples:
* `vigenere_encrypt('FunctION', 'JokE')` returns `OixgcWYR`
* `vigenere_encrypt('HappyBirthdayy', 'presENTS')` returns `WrthcObjiyhscl`
* `vigenere_encrypt('SpongeBobSquarePants', 'ImPrOvEtHeAlGoRiThM')` returns `AbdeuzFhiWqfgfvXtufa`

```
def vigenere_encrypt(plaintext, keyword):
    keyword = keyword * (len(plaintext) // len(keyword) + 1)
    plaintext_nums = [ord(ch.upper()) - ord('A') for ch in plaintext]
    keyword_nums = [ord(ch.upper()) - ord('A') for ch in keyword]
    ciphertext = ''
    for i in range(len(plaintext)):
        cipher = (plaintext_nums[i]+keyword_nums[i]) % 26
        ciphertext += chr(cipher + ord('A') if plaintext[i].isupper() else cipher + ord('a'))
    return ciphertext

print(vigenere_encrypt('FunctION', 'JokE'))
print(vigenere_encrypt('HappyBirthdayy', 'presENTS'))
print(vigenere_encrypt('SpongeBobSquarePants', 'ImPrOvEtHeAlGoRiThM'))
```
