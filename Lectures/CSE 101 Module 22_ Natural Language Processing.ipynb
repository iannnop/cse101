```
import re
import random

regexes = {
    r'^i (am|feel) (.+).*$': r'Why do you feel \2 today? ',
    r'^i lost my (.+)$': r'Hmm, unfortunate. How did you lose your \1? ',
    r'^i (remember|forget) (.+)$': r'What about \2 do you \1? ',
    r'^i (prefer|like|love|hate|adore|despise) (.+)$': r'What about \2 do you \1? ',
    r'^i (eat|drink) (.+)$': "Isn't that really expensive? ",
    r'^(yes|no|yeah|nope)$': 'I see. Go on. ',
}

# main loop
line = input('Hello. Tell me what\'s on your mind. ')
while line.lower() != 'bye':
    line = line.lower().strip()
    found_match = False
    for regex in regexes:
        if re.match(regex, line):
            line = input(re.sub(regex, regexes[regex], line))
            found_match = True
            break
    if not found_match:
        question = random.choice([
            'Interesting. Tell me more about this.',
            'And how does that make you feel?',
            'Why do you think that is?',
            'Let\'s talk about your childhood.',
            'What do you dream about?',
            'What else is on your mind today?',
        ])
        line = input(question + ' ')

print('Goodbye!')
```
