### Syntax Errors

A **syntax error** is a mistake in your code that violates the rules of the language, known as the language's **syntax**. Such errors prevent your code from running to completion.

The **Python interpreter**, the code inside Colab that "understands" Python, will alert you to the error and where to find it in your code.

### Runtime Errors

A runtime error arises only when you run a program and causes your program to crash. A common run-time error is a division-by-zero error:

```
numerator = 5
denominator = 0
fraction = numerator / denominator
fraction
```

### Logical Errors

A **logical error** also arises when you run a program and the program generates a result, but not the result you wanted. These kinds of errors can be hard to **debug** (fix).

```
item_price = 50
tax_rate = 0.8  # oops, we meant 0.08
tax = item_price * tax_rate
total = item_price + tax
total
```

The above result doesn't make sense...how could adding the tax almost double the price?

### Debugging

Debugging is the process of locating and fixing bugs in your code. Logical errors can be tricky to debug! One common way is to make use of print statements to display the *state* of variables during execution.

The print value for `total` doesn't make sense, so we should step back a few lines and start inserting print statements:

```
item_price = 50
tax_rate = 0.8  # oops, we meant 0.08
tax = item_price * tax_rate
print(f'tax = {tax}')
total = item_price + tax
total
```

By printing the tax computed, we discover that its value can't be right.

Here we are using an **f-string**, one way of printing the value of a variable or expression in Python. Whatever code you put inside the curly braces is evaluated and printed.

The `f` at the start of the string is required.
